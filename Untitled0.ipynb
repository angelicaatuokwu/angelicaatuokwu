```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import io
import zipfile

# The uploaded variable contains the content of 'archive.zip'
# We need to extract the CSV file from the zip archive
with zipfile.ZipFile(io.BytesIO(uploaded['archive.zip']), 'r') as z:
    # Assuming the CSV file inside the zip is 'most_subscribed_youtube_channels.csv'
    # You can use z.namelist() to see all files inside the zip if unsure.
    csv_file_name_in_zip = 'most_subscribed_youtube_channels.csv'

    with z.open(csv_file_name_in_zip) as csv_file:
        df = pd.read_csv(csv_file)
        display(df)
```

```python
import plotly.express as px
from IPython.display import display

# Clean the 'video count' column: remove commas and convert to numeric
df['video count'] = df['video count'].astype(str).str.replace(',', '', regex=False).astype(float)

fig = px.histogram(df, x='video count', title='Video Count Distribution')
display(fig)
```

```python
import plotly.express as px
from IPython.display import display

# Ensure 'subscribers' is numeric (already handled in a previous step, but good practice)
df['subscribers'] = df['subscribers'].astype(str).str.replace(',', '', regex=False).astype(float)

# Group by 'started' year and sum the 'subscribers'
subscribers_by_year = df.groupby('started')['subscribers'].sum().reset_index()

# Create a pie chart
fig = px.pie(
    subscribers_by_year,
    values='subscribers',
    names='started',
    title='Total Subscribers by Channel Start Year',
    labels={'started': 'Year Channel Started', 'subscribers': 'Total Subscribers'}
)

display(fig)
```

```python
import plotly.express as px

fig = px.box(df, y='started', title='Years Started')
fig.show()
```
